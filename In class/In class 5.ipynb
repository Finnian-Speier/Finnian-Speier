```python
def total(List):
    total = 0
    n = len(List)
    for i in range(n):
        total += List[i]
    return total
#this is a list i made to test some of the funtions 
job=[1,2,3,4,5,6,7,8,1,1,2,6,8,25,32]

print(total(job))

def mean(List):
    mean_ = total(List) / len(List)
    return mean_

print(mean(job))

def median(List):
    n = len(List)
    List = sorted(List)
    #figurues out if its odd or even, since taking the median is diffrent for even lists since there is no true middle
    if n % 2 != 0:
        middle_index = int((n - 1) / 2)
        median_ = List[middle_index]
    else:
        upper_middle_index = int(n / 2)
        lower_middle_index = (upper_middle_index - 1)
        median_ = mean(List[lower_middle_index : upper_middle_index + 1])
        
    return median_
print(median(job))
def mode(List):
    max_count = 0
    counter_dict = {}
    for value in List:
        counter_dict[value] = 0
    for value in List:
        #just adds one to that vales count for each instance of it in the list
        counter_dict[value] += 1
    count_list = list(counter_dict.values())
    #this line figures out witch of the numbers has the highest counts 
    max_count = max(count_list)
    mode_ = [key for key in counter_dict if counter_dict[key] == max_count]
    
    return mode_
print(mode(job))
def variance(List, sample = False):
    #for this funtion there needs to be two arguments one for the list and the other for weather its a sample or not
    #calles the mean funtion
    list_mean = mean(List)
    n = len(List)
    sum_sq_diff = 0
    for val in List:
        #this line takes all values in the list and subtracts them from the mean and squares the result
        sum_sq_diff += (val - list_mean) ** 2
    if sample == False:
        variance_ = sum_sq_diff / n
    else:
        variance_ = sum_sq_diff / (n - 1)
    
    return variance_
print()
def SD(List, sample = False):
    #for this funtion there needs to be two arguments one for the list and the other for weather its a sample or not the third agrument shoudl be a boolian
    SD_ = variance(List, sample) ** (1/2)
    
    return SD_
print(SD(job))
def covariance(List1, List2, sample = False):
    #this funtion requieres three agruments, one for each list and one for weather its a sample, the third agrument shoudl be a boolian
    mean1 = mean(List1)
    mean2 = mean(List2)
    cov = 0
    n1 = len(List1)
    n2 = len(List2)
    if n1 == n2:
        n = n1
        for i in range(n1):
            cov += (List1[i] - mean1) * (List2[i] - mean2)
        if sample == False:
            cov = cov / n
        else:
            cov = cov / (n - 1)
        return cov
    else:
        print("List lengths are not equal")
        print("List1:", n1)
        print("List2:", n2)
    

def correlation(List1, List2):
    #this funtion bassically just uses eairlier funtions together
    cov = covariance(List1, List2)
    SD1 = SD(List1)
    SD2 = SD(List2)
    corr = cov / (SD1 * SD2)
    return corr

def skewness(List, sample = False):
    #this funtion requieres sample to be faulse in order to run, idk if this was the right call achually
    mean_ = mean(List)
    SD_ = SD(List, sample)
    skew = 0
    n = len(List)
    for val in List:
        skew += (val - mean_) ** 3
    skew = skew / (n * SD_ **3) if not sample else n * skew / ((n - 1)*(n - 2) * SD_ ** 3)
        
    return skew

def kurtosis(List, sample = False):
    #this funtion requieres sample to be faulse in order to run, idk if this was the right call achually
    mean_ = mean(List)
    mean_ = mean(List)
    kurt = 0
    SD_ = SD(List, sample)
    n = len(List)
    for val in List:
        kurt += (val - mean_) ** 4
    kurt = kurt / (n * SD_ ** 4) if sample == False else  n * (n + 1) * kurt / \
    ( (n - 1) * (n - 2) *(n - 3) * (SD_ ** 4)) - (3 *(n - 1) ** 2) / ((n - 2) * (n - 3))
    
    return kurt



```

```python
def stats(df, sample = False):
    dct = {key:{} for key in df}
    for key, val in df.items():
        # drop any missing observations from dataframe
        val = val.dropna(axis=0)
        dct[key]["mean"] = round(mean(val),3)
        dct[key]["median"] = round(median(val),3)
        # skip mode. . .         dct[key]["mode"] = mode(val)
        dct[key]["variance"] = round(variance(val, sample),3)
        dct[key]["S.D."] = round(SD(val, sample) ,3)
        dct[key]["skewness"] = round(skewness(val, sample),3)
        dct[key]["kurtosis"] = round(kurtosis(val, sample),3)
    stats = pd.DataFrame(dct)
    return stats
```

```python
#our lovly imports 
import pandas as pd
import pandas_datareader.data as web
import datetime

#picking the start date
start = datetime.datetime(2010, 1 , 1)
#picking the end date
end = datetime.datetime.today()

data_dict = {}
#picking the stocks
stocks = [ "AAPL", "FB", "TSLA","HAS"]
for key in stocks:
    data_dict[key] = web.DataReader(key, 'yahoo', start, end)
```

```python
close_data = pd.concat([val["Close"] for val in data_dict.values()], keys = data_dict.keys(), axis = 1)
close_data
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 32})
fig, ax = plt.subplots(figsize = (24,16))
close_data.plot.line(ax = ax, legend = True)
y_vals = ax.get_yticks()
ax.set_yticklabels([int(y) if y >= 1 else round(y,1) for y in y_vals])    
#bitween 2010 and 2015 hasbron was competative with the rest of these but it got crushed after 2016

plt.title("Stock prices of Tesla Microsft Apple and Hasbro")
plt.show()
plt.close()
```
