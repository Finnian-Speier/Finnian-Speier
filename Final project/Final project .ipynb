    During the COVID-19 pandemic, many of us got haircuts from friends and family. Those who did will likely know that having your hair cut by a non-trained professional is not dangerous. While some of us got a bad haircut, few of us were injured. According to the CDC, while accidents are one of the leading causes of death, deaths involving scissors are an insignificant part of those deaths. On top of that haircuts aren't even the most common cause of scissor-related injuries, running with them is far more dangerous. 
	So why do you need a license to cut hair professionally? One doesn't need a license to cook and sell food, which according to the CDC causes 3,000 deaths a year and 48 million instances of food poisoning a year. If it's an issue of safety then an industry like cutting hair, which is far safer than cooking, should not require a license, and cooking food should. So then why does someone need a license to cut hair? This paper will aim to explain why licenses are in place in the United States and show their effect on the economy. 
    While few people know about the licensing process of hairdressers, their licenses are surprisingly difficult to obtain. These licenses aren't something you could get in a week, or in many cases even a year. This is a chart showing the difficulty to get a hairdressing license by the state. 


```python
import pandas as pd
import matplotlib.pyplot as plt
 
#C:\Users\The Grand Tabaxi\Github repository\Finnian-Speier\Final project

df=pd.DataFrame([1.31578947,3.94736842,9.118421053,40.789,32.8947368,9.21052632] , 
    index = ["Less than 800","Between 800 and 900","Between 900 and 1000","Between 1000 and 1500","1500","Greater then 1500"]) 
df.plot(kind="pie" , subplots="True" , figsize=(6,6))    
plt.legend(loc='lower right')
plt.title("Hours to get hairdressing license")
```

    This chart represents the hours it takes to get a haircut by state. As we can see the burden of becoming a hairdresser is massive with most states requiring more than a thousand hours of work to get a license. Total hours can include hours spent in a classroom or at a required apprenticeship. This also includes hours spent taking tests. These licenses are also not cheap; it can cost between $10,000 and $20,000 depending on the state you get them in. 
    When you include the cost of the lost time it can cost a prospective barber upwards of $55,000 to get their license, assuming one hour is worth their future average hourly rate of $23 an hour. However, aside from the magnitude of these costs, there is something else interesting about this data. The massive variance between states. 
    New York has the lowest requirements to become a barber with only 231 hours of training required.  Iowa dwarfs New York's requirements with a massive 2,100 required hours. Logically this should do a few things. First, fewer people should be willing to become hairdressers because the training requirement is so high. That would cause the supply of hairdressers to decrease which would, in turn, decrease the supply of haircuts, and since this wouldn't affect the demand for haircuts, it would cause the price of a haircut to increase. This is a well-documented economic phenomenon that has many papers written on it. 
    One such paper is Public Choice Theory and Occupational Licensing by Jiong Tu, which details the effects occupational licensing has on the economy. It demonstrates that as occupational licensing and other regulatory burdens increase so too does the price. This is because fewer people will want to go into the field, and as the paper pointed out those who do go into the field will spend less time working in it. 
    For example, say it takes someone two years to get a license and they would have worked for 40 years then 5 percent of their time is lost to the license. That means that there is effectively 5 percent less supply assuming the same number of people go into the field. While the theoretical effect this might have on price is impossible to calculate in practice it does increase price almost all the time. This paper does take a clear stance against occupational licensing saying that it removes individual freedom from people by preventing them from pursuing their goals without the government's blessing, and limits consumers' choices. However, the paper will present a neutral stance for objectivity.
    Given that New York has a much lower barrier to entry we would expect the cost of a haircut in New York to be much cheaper than one in Iowa, at least when we control for factors like cost of living. However, that couldn't be further from the truth. 
    The cost of a haircut in Iowa is only 18 dollars on average while in New York the average cost is a whopping $37. Even if you control for the cost of living, the average cost of a haircut in New York is still more expensive by 8 dollars. This could be an anomaly limited to New York and Iowa, perhaps the people of New York are simply willing to pay more for a haircut than the people of Iowa. Or it's possible that the more wealthy people of New York expect a higher quality haircut and are willing to pay more for it. However, the data shows that this is not an anomaly but a trend throughout the United States.
